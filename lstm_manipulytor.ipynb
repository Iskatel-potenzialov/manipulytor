```python
# @title Обработка 4 файлов CSV → один общий DataFrame + Графики
import pandas as pd
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
from scipy.ndimage import uniform_filter1d
import json

# ============================================================================
# НАСТРОЙКИ ДЛЯ 4 ФАЙЛОВ
# ============================================================================
FILES_CONFIG = [
    {'file': '/content/data_1_7.csv',   'n_tohki': 1,  'detect_pot': 1},  # pot2
    {'file': '/content/data_8_16.csv',  'n_tohki': 8,  'detect_pot': 3},  # pot4
    {'file': '/content/data_17_25.csv', 'n_tohki': 17, 'detect_pot': 3},  # pot4
    {'file': '/content/data_26_34.csv', 'n_tohki': 26, 'detect_pot': 3},  # pot4
]

TARGET_LENGTH = 200
PAUSE_THRESHOLD = 3
MIN_PAUSE_DURATION = 1000
MIN_MOVEMENT_DURATION = 30
DETECT_SMOOTH_WINDOW = 100
LSTM_SMOOTH_WINDOW = 3

# ============================================================================
# ОБЩИЕ КОНТЕЙНЕРЫ ДЛЯ ВСЕХ ФАЙЛОВ
# ============================================================================
all_trajectories = []
all_trajectory_info = []
all_df_list = []

# ============================================================================
# ЦИКЛ ПО ВСЕМ ФАЙЛАМ
# ============================================================================
for file_idx, config in enumerate(FILES_CONFIG):
    FILE_NAME = config['file']
    n_tohki = config['n_tohki']
    DETECT_POT_INDEX = config['detect_pot']

    print("\n" + "=" * 80)
    print(f"📁 ФАЙЛ {file_idx + 1}/{len(FILES_CONFIG)}: {FILE_NAME}")
    print(f"   n_tohki = {n_tohki}, DETECT_POT_INDEX = {DETECT_POT_INDEX}")
    print("=" * 80)

    # ============================================================================
    # ЗАГРУЗКА ДАННЫХ
    # ============================================================================
    try:
        df = pd.read_csv(FILE_NAME)
        df.columns = df.columns.str.strip()
    except Exception as e:
        print(f"❌ Ошибка загрузки {FILE_NAME}: {e}")
        continue

    pot_columns = ['pot1', 'pot2', 'pot3', 'pot4']
    missing_cols = [col for col in pot_columns if col not in df.columns]
    if missing_cols:
        print(f"❌ Ошибка: не найдены колонки {missing_cols}")
        continue

    data = df[pot_columns].values
    time = df['time_ms'].values

    print(f"✅ Загружено {len(data)} строк ({len(data)/1000:.1f} сек)")

    # ============================================================================
    # 1. ДЕТЕКЦИЯ ПАУЗ + ВИЗУАЛИЗАЦИЯ
    # ============================================================================
    print(f"\n🔍 ДЕТЕКЦИЯ ПАУЗ:")
    print(f"   Используем: {pot_columns[DETECT_POT_INDEX]}")
    print(f"   Окно сглаживания: {DETECT_SMOOTH_WINDOW}")
    print("=" * 80)

    detect_data = data[:, DETECT_POT_INDEX]
    detect_smooth = uniform_filter1d(detect_data, size=DETECT_SMOOTH_WINDOW, axis=0)

    detect_window = 100
    variance = np.zeros(len(data))

    for i in range(detect_window//2, len(data) - detect_window//2):
        window = detect_smooth[i-detect_window//2 : i+detect_window//2]
        variance[i] = np.std(window)

    is_pause = variance < PAUSE_THRESHOLD

    pause_segments = []
    in_pause = is_pause[detect_window//2]
    pause_start = detect_window//2

    for i in range(detect_window//2 + 1, len(is_pause)):
        if is_pause[i] != in_pause:
            if in_pause and (i - pause_start) >= MIN_PAUSE_DURATION:
                pause_segments.append({
                    'start': int(pause_start),
                    'end': int(i),
                    'length': int(i - pause_start),
                    'center': int((pause_start + i) // 2)
                })
            pause_start = i
            in_pause = is_pause[i]

    if in_pause and (len(is_pause) - pause_start) >= MIN_PAUSE_DURATION:
        final_end = min(len(time) - 1, len(is_pause) - 1)
        pause_segments.append({
            'start': int(pause_start),
            'end': int(final_end),
            'length': int(final_end - pause_start),
            'center': int((pause_start + final_end) // 2)
        })

    print(f"📍 Найдено пауз: {len(pause_segments)}")

    # ← ВИЗУАЛИЗАЦИЯ ПАУЗ
    plt.figure(figsize=(16, 6))
    plt.plot(time, detect_data, label=f'{pot_columns[DETECT_POT_INDEX]} (сырые)', alpha=0.5, linewidth=0.5)
    plt.plot(time, detect_smooth, label=f'{pot_columns[DETECT_POT_INDEX]} (smooth win={DETECT_SMOOTH_WINDOW})', linewidth=2, color='green')

    for i, pause in enumerate(pause_segments):
        start_idx = min(pause['start'], len(time) - 1)
        end_idx = min(pause['end'], len(time) - 1)
        plt.axvspan(time[start_idx], time[end_idx],
                   alpha=0.2, color='green', label='Пауза' if i==0 else '')

    plt.title(f'Файл {file_idx+1}: Детекция пауз по {pot_columns[DETECT_POT_INDEX]} (зелёные зоны = паузы)', fontsize=12, fontweight='bold')
    plt.xlabel('Время (ms)')
    plt.ylabel('Значение потенциометра')
    plt.legend(loc='upper right', fontsize=9)
    plt.grid(True, alpha=0.3)

    pause_plot_filename = f'pause_detection_file{file_idx+1}.png'
    plt.savefig(pause_plot_filename, dpi=150, bbox_inches='tight')
    print(f"   Сохранено: {pause_plot_filename}")
    plt.show()

    # ============================================================================
    # 2. ИЗВЛЕЧЕНИЕ ТРАЕКТОРИЙ
    # ============================================================================
    print(f"\n🤖 ИЗВЛЕЧЕНИЕ ТРАЕКТОРИЙ ДЛЯ LSTM:")
    print(f"   Окно сглаживания: {LSTM_SMOOTH_WINDOW}")
    print("=" * 80)

    data_lstm = uniform_filter1d(data, size=LSTM_SMOOTH_WINDOW, axis=0)


    local_point_num = 0

    for i in range(len(pause_segments) - 1):
        pause1 = pause_segments[i]
        pause2 = pause_segments[i + 1]

        move_start = max(0, pause1['end'])
        move_end = min(len(data_lstm), pause2['start'])

        if move_start >= move_end:
            continue

        move_length = move_end - move_start
        if move_length < MIN_MOVEMENT_DURATION:
            continue

        traj = data_lstm[move_start:move_end]
        traj_normalized = traj / 1023.0

        original_length = len(traj_normalized)
        x_old = np.linspace(0, 1, original_length)
        x_new = np.linspace(0, 1, TARGET_LENGTH)

        resampled = np.zeros((TARGET_LENGTH, 4))
        for joint in range(4):
            f = interp1d(x_old, traj_normalized[:, joint], kind='linear')
            resampled[:, joint] = f(x_new)

        all_trajectories.append(resampled)

        # ← Определяем направление
        is_even_trajectory = len(all_trajectories) % 2 == 0
        direction = 'TO_BASE' if is_even_trajectory else 'TO_POINT'


        if direction == 'TO_POINT':
            local_point_num += 1  # ← Увеличиваем ТОЛЬКО для TO_POINT
            current_tohka = n_tohki + local_point_num - 1
            point_num = current_tohka  # ← point_num = tohka для TO_POINT
        else:
            current_tohka = 0  # ← База всегда 0
            point_num = local_point_num  # ← Для TO_BASE берём последний local_point_num

        all_trajectory_info.append({
            'index': int(len(all_trajectories)),
            'file_idx': file_idx,
            'file_name': FILE_NAME,
            'start_time': int(time[move_start]),
            'end_time': int(time[move_end - 1]),
            'duration_ms': int(time[move_end - 1] - time[move_start]),
            'original_points': int(original_length),
            'direction': direction,
            'point_num': int(point_num),
            'tohka': int(current_tohka),
            'from_pause': int(i),
            'to_pause': int(i + 1)
        })

        # Добавляем в DataFrame список
        for step in range(len(resampled)):
            row = {
                'trajectory_id': len(all_trajectories) - 1,
                'file_idx': file_idx,
                'file_name': FILE_NAME.split('/')[-1],
                'tohka': current_tohka,
                'step': step,
                'pot1': resampled[step, 0],
                'pot2': resampled[step, 1],
                'pot3': resampled[step, 2],
                'pot4': resampled[step, 3],
                'direction': direction,
                'point_num': point_num,
                'duration_ms': int(time[move_end - 1] - time[move_start]),
                'original_points': original_length
            }
            all_df_list.append(row)

        print(f"  Траектория {len(all_trajectories):2d}: {direction} (Точка #{point_num}, tohka={current_tohka})")

    print(f"✅ Файл {FILE_NAME} обработан. Всего траекторий: {len(all_trajectories)}")
    print(f"   Точек в этом файле: {local_point_num} (tohka {n_tohki} → {n_tohki + local_point_num - 1})")

# ============================================================================
# 3. СОЗДАНИЕ ОБЩЕГО DATAFRAME
# ============================================================================
print("\n" + "=" * 80)
print("📋 СОЗДАНИЕ ОБЩЕГО DATAFRAME ДЛЯ ВСЕХ ФАЙЛОВ")
print("=" * 80)

df_trajectories = pd.DataFrame(all_df_list)
print(f"✅ DataFrame создан: {df_trajectories.shape}")
print(f"   Колонки: {list(df_trajectories.columns)}")

print(f"\n📊 ПЕРВЫЕ 10 СТРОК:")
print(df_trajectories.head(10).to_string(index=False))

print(f"\n📈 СТАТИСТИКА ПО ФАЙЛАМ:")
stats_by_file = df_trajectories.groupby('file_name').agg({
    'trajectory_id': 'nunique',
    'step': 'count'
}).rename(columns={'trajectory_id': 'trajectories', 'step': 'total_steps'})
print(stats_by_file)

print(f"\n📈 СТАТИСТИКА ПО TOHKA (первые 10):")
stats_by_tohka = df_trajectories.groupby('tohka').agg({
    'trajectory_id': 'nunique',
    'step': 'count'
}).rename(columns={'trajectory_id': 'trajectories', 'step': 'total_steps'})
print(stats_by_tohka.head(10))

# Сохранение DataFrame
df_trajectories.to_csv('lstm_trajectories_combined_debug.csv', index=False)
print(f"\n💾 DataFrame сохранён в: lstm_trajectories_combined_debug.csv")

# ============================================================================
# 4. СОХРАНЕНИЕ .NPY И .JSON
# ============================================================================
dataset = np.array(all_trajectories)
np.save('lstm_trajectories_combined.npy', dataset)

print(f"\n✅ Датасет сохранен: lstm_trajectories_combined.npy")
print(f"   Форма: {dataset.shape}")
print(f"   [кол-во траекторий, {TARGET_LENGTH} шагов, 4 сустава]")

with open('trajectory_info_combined.json', 'w', encoding='utf-8') as f:
    json.dump(all_trajectory_info, f, indent=2, ensure_ascii=False)
print(f"   Метаданные: trajectory_info_combined.json")

# ============================================================================
# 5. ВИЗУАЛИЗАЦИЯ ПАРАМИ (БАЗА→ТОЧКА | ТОЧКА→БАЗА)
# ============================================================================
pairs = []
for i in range(0, len(all_trajectories) - 1, 2):
    pairs.append({
        'to_point_idx': i,
        'to_base_idx': i + 1,
        'point_num': all_trajectory_info[i]['point_num'],
        'tohka': all_trajectory_info[i]['tohka']
    })

if len(all_trajectories) % 2 != 0:
    print(f"\n⚠️ Последняя траектория #{len(all_trajectories)} без пары (нет возврата в базу)")

pairs_per_fig = 4
num_figures = max(1, (len(pairs) + pairs_per_fig - 1) // pairs_per_fig)

print(f"\n📈 Генерация {num_figures} фигур с графиками траекторий...")

for fig_num in range(num_figures):
    start_pair = fig_num * pairs_per_fig
    end_pair = min(start_pair + pairs_per_fig, len(pairs))

    fig, axes = plt.subplots(pairs_per_fig, 2, figsize=(16, 12))

    for row_idx, pair_idx in enumerate(range(start_pair, end_pair)):
        pair = pairs[pair_idx]
        point_num = pair['point_num']
        tohka = pair['tohka']

        # ЛЕВАЯ КОЛОНКА: БАЗА → ТОЧКА
        ax_left = axes[row_idx, 0]
        traj_to = all_trajectories[pair['to_point_idx']]
        info_to = all_trajectory_info[pair['to_point_idx']]

        colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
        for joint in range(4):
            ax_left.plot(np.linspace(0, 1, TARGET_LENGTH), traj_to[:, joint],
                       label=f'Pot{joint+1}', linewidth=2, color=colors[joint])

        ax_left.set_title(f'Tohka #{tohka}: БАЗА → ТОЧКА\n{info_to["duration_ms"]} мс',
                        fontsize=11, fontweight='bold')
        ax_left.set_xlabel('Прогресс (0=старт, 1=финиш)')
        ax_left.set_ylabel('Нормализованное значение [0-1]')
        ax_left.legend(loc='upper right', fontsize=8)
        ax_left.grid(True, alpha=0.3)
        ax_left.set_ylim(-0.05, 1.05)

        # ПРАВАЯ КОЛОНКА: ТОЧКА → БАЗА
        ax_right = axes[row_idx, 1]
        traj_back = all_trajectories[pair['to_base_idx']]
        info_back = all_trajectory_info[pair['to_base_idx']]

        for joint in range(4):
            ax_right.plot(np.linspace(0, 1, TARGET_LENGTH), traj_back[:, joint],
                        label=f'Pot{joint+1}', linewidth=2, color=colors[joint])

        ax_right.set_title(f'Tohka #{tohka}: ТОЧКА → БАЗА\n{info_back["duration_ms"]} мс',
                         fontsize=11, fontweight='bold')
        ax_right.set_xlabel('Прогресс (0=старт, 1=финиш)')
        ax_right.set_ylabel('Нормализованное значение [0-1]')
        ax_right.legend(loc='upper right', fontsize=8)
        ax_right.grid(True, alpha=0.3)
        ax_right.set_ylim(-0.05, 1.05)

    # Скрываем пустые subplot
    for row_idx in range(end_pair - start_pair, pairs_per_fig):
        fig.delaxes(axes[row_idx, 0])
        fig.delaxes(axes[row_idx, 1])

    plt.suptitle(f'Траектории манипулятора (Пары {start_pair + 1}-{end_pair})',
                fontsize=14, fontweight='bold', y=0.98)
    plt.tight_layout(rect=[0, 0, 1, 0.96])

    output_filename = f'trajectories_pairs_combined_fig{fig_num + 1}.png'
    plt.savefig(output_filename, dpi=150, bbox_inches='tight')
    print(f"   Сохранено: {output_filename}")
    plt.show()

# ============================================================================
# 6. ИТОГОВАЯ СТАТИСТИКА
# ============================================================================
print("\n" + "=" * 80)
print("📊 ИТОГОВАЯ СТАТИСТИКА:")
print("=" * 80)
print(f"   Обработано файлов: {len(FILES_CONFIG)}")
print(f"   Всего траекторий: {len(all_trajectories)}")
print(f"   Всего точек (пар): {len(pairs)}")
print(f"   Обучающих пар (×199): ~{len(all_trajectories) * 199:,}")

durations = [t['duration_ms'] for t in all_trajectory_info]
to_point_durations = [t['duration_ms'] for t in all_trajectory_info if t['direction'] == 'TO_POINT']
to_base_durations = [t['duration_ms'] for t in all_trajectory_info if t['direction'] == 'TO_BASE']

print(f"\n   Длительность (все траектории):")
print(f"      Средняя: {np.mean(durations):.0f} мс")
print(f"      Мин/Макс: {min(durations):.0f} / {max(durations):.0f} мс")

print(f"\n   Длительность (БАЗА → ТОЧКА):")
if to_point_durations:
    print(f"      Средняя: {np.mean(to_point_durations):.0f} мс")

print(f"\n   Длительность (ТОЧКА → БАЗА):")
if to_base_durations:
    print(f"      Средняя: {np.mean(to_base_durations):.0f} мс")

print(f"\n✅ Обработка всех файлов завершена успешно!")
```

```python
# @title Функция перевода из нормализованных значений потенциометра в углы (градусы)
def norm_to_angle(norm):
    """
    norm: список из 4 нормализованных значений (0-1)
    возвращает список из 4 углов в градусах
    """
    raw = [v * 1023 for v in norm]
    params = [
        (523, 3.81),  # pot1: const, coeff
        (869, 3.65),  # pot2
        (740, 3.93),  # pot3
        (963, 3.89)   # pot4
    ]
    angles = [(const - r) / coeff for r, (const, coeff) in zip(raw, params)]
    return angles

# Примеры использования
points = {
    "база": [0.264735, 0.434190, 0.366425, 0.499252],
    "точка 1": [0.16617791, 0.63147605, 0.60312805, 0.47898338],
    "точка 27":  [0.10459433, 0.4740958, 0.3255132, 0.57575758],
    "точка 23": [0.35190616, 0.48289345, 0.34408602, 0.55816227],
    "точка 33": [0.40860215, 0.45454545, 0.31182796, 0.56793744],
    "точка 4":[0.25513196, 0.55034213, 0.42815249, 0.55131965],
    "точка 26":[0.0684262,  0.50928641, 0.3773216,  0.55913978],
    "точка 5":[0.2864, 0.5523, 0.4194, 0.5572],
    "точка 15":[0.35288368, 0.50733138, 0.42913001, 0.54740958],
}
for name, norm in points.items():
    angles = norm_to_angle(norm)
    print(f"{name}: {[round(a, 2) for a in angles]}°")
```

```python
# @title ! СОБИРАЕМ X и y из df_trajectories
X_list = []
y_list = []
weight_list = []  # дополнительный список для весов

TARGET_LENGTH = 200
alpha = 2  # коэффициент усиления веса на последнем шаге

print(f"📊 Обрабатываю DataFrame: {df_trajectories.shape}")

for traj_id, group in df_trajectories.groupby('trajectory_id'):
    group = group.sort_values('step').reset_index(drop=True)
    target_angles = group[['pot1', 'pot2', 'pot3', 'pot4']].iloc[-1].values

    for i in range(len(group) - 1):
        current_angles = group[['pot1', 'pot2', 'pot3', 'pot4']].iloc[i].values
        next_angles = group[['pot1', 'pot2', 'pot3', 'pot4']].iloc[i + 1].values


        step = group.iloc[i]['step']
        progress = step / (TARGET_LENGTH - 1)

        # Вес: линейно растёт от 1 до alpha
        weight = 1.0 + (alpha - 1.0) * (step / (TARGET_LENGTH - 1))

        x_input = np.concatenate([current_angles, target_angles, [progress]])
        y_output = next_angles

        X_list.append(x_input)
        y_list.append(y_output)
        weight_list.append(weight)

X = np.array(X_list)          # (M, 9)
y = np.array(y_list)          # (M, 4)
weights = np.array(weight_list)  # (M,)

print(f"\n✅ Обучающие данные готовы:")
print(f"   X shape: {X.shape}")
print(f"   y shape: {y.shape}")
print(f"   Веса: min={weights.min():.3f}, max={weights.max():.3f}")

np.save('lstm_X_progress_weighted.npy', X)
np.save('lstm_y_progress_weighted.npy', y)
np.save('lstm_weights_progress_weighted.npy', weights)
print("   Сохранено: lstm_X_progress_weighted.npy, lstm_y_progress_weighted.npy, lstm_weights_progress_weighted.npy")
```

```python
# @title ! Обучение LSTM + график обучения на 512 и 300 эпох
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
import json
import matplotlib.pyplot as plt

# Загружаем данные с весами
weights = np.load('lstm_weights_progress_weighted.npy')

# Преобразуем в тензоры
X_tensor = torch.tensor(X, dtype=torch.float32).unsqueeze(1)  # (M, 1, 9)
y_tensor = torch.tensor(y, dtype=torch.float32)               # (M, 4)
weights_tensor = torch.tensor(weights, dtype=torch.float32)    # (M,)

print(f"Всего примеров: {len(X_tensor)}")

# Разбиение на train/val/test
total_len = len(X_tensor)
train_len = int(0.7 * total_len)
val_len = int(0.15 * total_len)

indices = torch.randperm(total_len)
X_shuffled = X_tensor[indices]
y_shuffled = y_tensor[indices]
w_shuffled = weights_tensor[indices]

X_train, y_train, w_train = X_shuffled[:train_len], y_shuffled[:train_len], w_shuffled[:train_len]
X_val, y_val, w_val = X_shuffled[train_len:train_len+val_len], y_shuffled[train_len:train_len+val_len], w_shuffled[train_len:train_len+val_len]
X_test, y_test, w_test = X_shuffled[train_len+val_len:], y_shuffled[train_len+val_len:], w_shuffled[train_len+val_len:]

# DataLoader (батчи с весами)
train_dataset = TensorDataset(X_train, y_train, w_train)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)

# Модель (input_size=9)
class TrajectoryLSTM(nn.Module):
    def __init__(self, input_size=9, hidden_size=512, num_layers=2, output_size=4, dropout=0.2):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers,
                           batch_first=True, dropout=dropout if num_layers > 1 else 0)
        self.fc = nn.Linear(hidden_size, output_size)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        lstm_out = lstm_out[:, -1, :]
        lstm_out = self.dropout(lstm_out)
        return self.fc(lstm_out)

model = TrajectoryLSTM()
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Взвешенная MSE: используем nn.MSELoss(reduction='none') и умножаем на веса
def weighted_mse_loss(pred, target, weight):
    loss = nn.MSELoss(reduction='none')(pred, target)  # (batch, 4)
    loss = loss.mean(dim=1)  # усредняем по суставам → (batch,)
    return (loss * weight).mean()  # взвешенное среднее

optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=10, factor=0.5)

epochs = 300 #300
best_val_loss = float('inf')
train_losses, val_losses = [], []

for epoch in range(epochs):
    model.train()
    train_loss = 0
    for batch_x, batch_y, batch_w in train_loader:
        batch_x, batch_y, batch_w = batch_x.to(device), batch_y.to(device), batch_w.to(device)
        optimizer.zero_grad()
        pred = model(batch_x)
        loss = weighted_mse_loss(pred, batch_y, batch_w)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()
    train_loss /= len(train_loader)
    train_losses.append(train_loss)

    model.eval()
    with torch.no_grad():
        val_pred = model(X_val.to(device))
        val_loss = weighted_mse_loss(val_pred, y_val.to(device), w_val.to(device)).item()
    val_losses.append(val_loss)

    scheduler.step(val_loss)

    if val_loss < best_val_loss:
        best_val_loss = val_loss
        torch.save(model.state_dict(), 'lstm_trajectory_best_weighted.pth')
        best_epoch = epoch + 1  # запоминаем лучшую эпоху

    if (epoch+1) % 10 == 0:
        print(f"Epoch {epoch+1:3d} | Train Loss: {train_loss:.6f} | Val Loss: {val_loss:.6f}")

# Тестирование
model.load_state_dict(torch.load('lstm_trajectory_best_weighted.pth'))
model.eval()
with torch.no_grad():
    test_pred = model(X_test.to(device))
    test_loss = weighted_mse_loss(test_pred, y_test.to(device), w_test.to(device)).item()
print(f"\n✅ Test Loss: {test_loss:.6f} (best epoch: {best_epoch})")

# ========================= ДОБАВЛЯЕМ ГРАФИК ОБУЧЕНИЯ =========================
plt.figure(figsize=(10, 5))
plt.plot(train_losses, label='Train Loss', alpha=0.7, linewidth=2)
plt.plot(val_losses, label='Val Loss', alpha=0.7, linewidth=2)
plt.axhline(y=best_val_loss, color='green', linestyle='--', label=f'Best Val: {best_val_loss:.6f}')
plt.xlabel('Epoch')
plt.ylabel('Weighted MSE Loss')
plt.title('Обучение LSTM с прогрессом и весами')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('training_loss_weighted.png', dpi=150)
plt.show()
print("   ✅ График сохранён: training_loss_weighted.png")
# =============================================================================

# Сохраняем параметры нормализации (добавляем best_epoch)
alpha = 2.0  # значение, использованное при генерации весов
norm_params = {
    'input_size': 9,
    'hidden_size': 128,
    'num_layers': 2,
    'output_size': 4,
    'best_val_loss': best_val_loss,
    'best_epoch': best_epoch,
    'test_loss': test_loss,
    'weight_alpha': alpha
}
with open('lstm_normalization_params_weighted.json', 'w') as f:
    json.dump(norm_params, f, indent=2)
print("   ✅ Параметры сохранены в lstm_normalization_params_weighted.json")
```
